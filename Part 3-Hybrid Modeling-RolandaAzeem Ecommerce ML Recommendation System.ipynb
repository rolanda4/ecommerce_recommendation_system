```python
#building hierarchy for the categories to better understand relationships when learning

def get_ancestors(cat_id, tree_dict):
    path = []
    while cat_id in tree_dict and pd.notna(tree_dict[cat_id]):
        cat_id = tree_dict[cat_id]
        path.append(cat_id)
    return path

# Build the mapping from child to parent
tree_dict = dict(zip(dup_category_tree["categoryid"], dup_category_tree["parentid"]))

# Compute the list of ancestors for each categoryid
dup_category_tree["ancestors"] = dup_category_tree["categoryid"].apply(lambda x: get_ancestors(x, tree_dict))
```

```python
# Build sparse matrices (using the mapped indices!)
from scipy.sparse import coo_matrix

train_csr = coo_matrix(
    (train_df['interaction_strength'],
     (train_df['user_idx'], train_df['item_idx']))
)

test_csr = coo_matrix(
    (test_df['interaction_strength'],
     (test_df['user_idx'], test_df['item_idx']))
)
```

```python
#to start, i will reweight the matrix

# Encode users and items
user_mapping = {u: i for i, u in enumerate(event_interaction_df['visitorid'].unique())}
item_mapping = {i: j for j, i in enumerate(event_interaction_df['itemid'].unique())}

event_interaction_df['user_idx'] = event_interaction_df['visitorid'].map(user_mapping)
event_interaction_df['item_idx'] = event_interaction_df['itemid'].map(item_mapping)

# Build user-item sparse matrix
user_item_csr = csr_matrix(
    (event_interaction_df['interaction_strength'],
     (event_interaction_df['user_idx'], event_interaction_df['item_idx']))
)
```

```python
from implicit.nearest_neighbours import bm25_weight

train_csr = train_csr.tocsr()  # make sure it's CSR

# --- BM25 weighting ---
bm25_train = bm25_weight(train_csr, K1=100, B=0.8)  # K1, B are tunable
bm25_train = bm25_train.tocsr()
```

```python
import random

all_users = list(range(user_item_csr.shape[0]))
random.seed(42)
valid_users = random.sample(all_users, int(0.2 * len(all_users)))  # 20% users for validation
```

```python
test_truth = {}
for u in valid_users:
    items = user_item_csr[u].indices
    if len(items) > 0:
        test_truth[u] = set(items)
```

```python
from implicit.als import AlternatingLeastSquares
als_tuned = AlternatingLeastSquares(
    factors=200,         
    regularization=0.01,
    iterations=20,
    random_state=42
)

als_tuned.fit(bm25_train)
```

```python
def precision_at_k(model, user_item_csr, test_truth, k=10):
    precisions = []
    num_users = user_item_csr.shape[0]

    for user, true_items in test_truth.items():
        if user >= num_users:  # skip unseen users
            continue

        recommended, _ = model.recommend(
            userid=user,
            user_items=user_item_csr[user],
            N=k,
            filter_already_liked_items=True,
            recalculate_user=False
        )

        hit_count = len(set(recommended) & true_items)
        precisions.append(hit_count / k)

    return np.mean(precisions) if precisions else 0.0


def map_at_k(model, user_item_csr, test_truth, k=10):
    aps = []
    num_users = user_item_csr.shape[0]

    for user, true_items in test_truth.items():
        if user >= num_users:  # skip unseen users
            continue

        recommended, _ = model.recommend(
            userid=user,
            user_items=user_item_csr[user],
            N=k,
            filter_already_liked_items=True,
            recalculate_user=False
        )

        if not true_items:
            continue

        score = 0.0
        hits = 0
        for i, item in enumerate(recommended, start=1):
            if item in true_items:
                hits += 1
                score += hits / i

        if hits > 0:
            aps.append(score / min(len(true_items), k))

    return np.mean(aps) if aps else 0.0
```

```python
prec10 = precision_at_k(als_tuned, bm25_train, test_truth, k=10)
map10 = map_at_k(als_tuned, bm25_train, test_truth, k=10)

print(f"weightPrecision@10: {prec10:.4f}")
print(f"weightMAP@10: {map10:.4f}")
```

#### Bringing it all together

```python
def hybrid_recommend(user, model, user_item_csr, N=10, 
                     alpha=0.6, beta=0.3, gamma=0.1,
                     item_popularity=None, item_similarity=None):
    
    # 1. ALS recommendations
    als_recs, als_scores = model.recommend(
        userid=user,
        user_items=user_item_csr[user],
        N=N*2,  # take more for blending
        filter_already_liked_items=True,
        recalculate_user=False
    )
    als_dict = {item: alpha * score for item, score in zip(als_recs, als_scores)}
    
    # 2. Popularity scores 
    pop_dict = {}
    if item_popularity is not None:
        for item, score in item_popularity.items():
            pop_dict[item] = beta * score
    
    # 3. Metadata similarity
    meta_dict = {}
    if item_similarity is not None:
        for item, score in item_similarity.items():
            meta_dict[item] = gamma * score
    
    # 4. Merge all scores
    final_scores = {}
    for d in [als_dict, pop_dict, meta_dict]:
        for item, score in d.items():
            final_scores[item] = final_scores.get(item, 0) + score
    
    # 5. Sort by score
    ranked = sorted(final_scores.items(), key=lambda x: -x[1])
    
    return [item for item, _ in ranked[:N]]
```

```python
# Example Usage
user_idx = 42  # replace with any user index from your dataset
print("Hybrid Recommendations for user", user_idx, ":", hybrid_recommend(user_idx, N=10))
```

#### Evaluating the hybrid model

```python
from collections import defaultdict

def evaluate_hybrid(test_users, test_truth, N=10):
    precisions, maps = [], []

    for user in test_users:
        # get hybrid recommendations
        recs = hybrid_recommend(user, N=N)

        # ground-truth items for this user (from test set)
        truth_items = test_truth.get(user, [])

        if not truth_items:  # skip if no ground truth
            continue

        # ---- Precision@K ----
        hits = sum(1 for item in recs if item in truth_items)
        precision = hits / N
        precisions.append(precision)

        # ---- MAP@K ----
        avg_prec, hit_count = 0, 0
        for rank, item in enumerate(recs, start=1):
            if item in truth_items:
                hit_count += 1
                avg_prec += hit_count / rank
        if truth_items:
            avg_prec /= min(len(truth_items), N)
        maps.append(avg_prec)

    return np.mean(precisions), np.mean(maps)
```

```python
# test_users is usually test_truth.keys()
prec, mapk = evaluate_hybrid(test_users=test_truth.keys(),
                             test_truth=test_truth,
                             N=10)

print(f"Hybrid Precision@10 = {prec:.4f}")
print(f"Hybrid MAP@10       = {mapk:.4f}")
```
